```python
Detect and Handle outliers from the dataset :# Fuction
```

```python
def handle_outliers(df, column, method="std", strategy="trim"):
    data = df[column]
    if method == "std":
        mean, std_dev = np.mean(data), np.std(data)
        lower, upper = mean - 3 * std_dev, mean + 3 * std_dev
    elif method == "percentile":
        lower, upper = np.percentile(data, [1, 99])
    elif method == "iqr":
        Q1, Q3 = np.percentile(data, [25, 75])
        IQR = Q3 - Q1
        lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
    elif method == "zscore":
        z_scores = zscore(data)
        mask = np.abs(z_scores) < 3
        if strategy == "trim":
            return df[mask]
        elif strategy == "cap":
            df[column] = np.where(mask, data, np.median(data))
            return df
    else:
        raise ValueError("Invalid method. Choose from 'std', 'percentile', 'iqr', or 'zscore'")
    
    if strategy == "trim":
        return df[(data >= lower) & (data <= upper)]
    elif strategy == "cap":
        df[column] = np.clip(data, lower, upper)
        return df
    elif strategy == "impute":
        df[column] = np.where((data < lower) | (data > upper), np.median(data), data)
        return df
    else:
        raise ValueError("Invalid strategy. Choose from 'trim', 'cap', or 'impute'")
```

```python
Applying all the methods to the dataset :- Mean and Standard Deviation, Percentile, IQR method ,Z-Score method.
```

```python
column = "price_per_sqft"  
df_std = handle_outliers(df.copy(), column, method="std", strategy="trim")
df_percentile = handle_outliers(df.copy(), column, method="percentile", strategy="trim")
df_iqr = handle_outliers(df.copy(), column, method="iqr", strategy="trim")
df_zscore = handle_outliers(df.copy(), column, method="zscore", strategy="trim")
```

```python
Boxplot Chart to compare all the method:
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=[df[column], df_std[column], df_percentile[column], df_iqr[column], df_zscore[column]])
plt.xticks([0, 1, 2, 3, 4], ["Original", "STD", "Percentile", "IQR", "Z-Score"])
plt.title("Comparison of Outlier Removal Methods")
plt.show()
```
